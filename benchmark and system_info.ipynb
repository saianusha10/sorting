```python
def insertionsort(array):
    for i in range(1, len(array)):
        k = array[i]
        j = i - 1
        while j >= 0 and k < array[j]:
           array[j + 1] = array[j]
           j -= 1
           array[j + 1] = k
    return array

a = insertionsort([2,6,4,9,1])
print(a)

```

```python
def selectionsort(array):
    for i in range(len(array)):
        min = i
        for j in range(i + 1, len(array)):
            if array[j] < array[min]:
                min = j
        array[i], array[min] = array[min], array[i]
    return array

a = selectionsort([23,12,20,50,60])
print(a)
```

```python
def bubblesort(array):
    n = len(array)
    for i in range(n):
        swap = False
        for j in range(0, n - i - 1):
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]
                swap = True
        if not swap:
            break
    return array

a = bubblesort([23,12,20,50,60])
print(a)
```

```python
import timeit
import numpy as np
import matplotlib.pyplot as plt

def benchmark(func, sizes):
    times = []
    for size in sizes:
        arr = np.random.randint(0, 1000, size)
        start_time = timeit.default_timer()
        func(arr.copy())
        end_time = timeit.default_timer()
        times.append(end_time - start_time)
    return times

sizes = [5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000]
insertion_times = benchmark(insertionsort, sizes)
selection_times = benchmark(selectionsort, sizes)
bubble_times = benchmark(bubblesort, sizes)

plt.plot(sizes, insertion_times, label='Insertion Sort')
plt.plot(sizes, selection_times, label='Selection Sort')
plt.plot(sizes, bubble_times, label='Bubble Sort')
plt.xlabel('Input size (n)')
plt.ylabel('Time (seconds)')
plt.legend()
plt.show()
```
